## Plot

Изобразим график функции, вычисленной на сетке, которая делит отрезок $[-1,1]$ на 10 равных частей.

Функция:
$$y=x(x+1)(x-1)$$

```python
X = np.linspace(-1., 1., 11)
print(X)
Y = X * (X+1) * (X-1)
plt.title('График функции')
plt.plot(X, Y, marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(1)
plt.minorticks_on()
plt.show()
```

```python
X = np.linspace(-1., 1., 101)
Y = X * (X+1) * (X-1)
plt.title('График функции')
plt.plot(X, Y, label=r'$y=x(x+1)(x-1)$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(1)
plt.minorticks_on()
plt.legend()
plt.show()
```

```python
X = np.linspace(-1., 1., 11)
Y = X * (X+1) * (X-1)
plt.plot(X, Y, marker='o')
plt.title('График функции')
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks([-1,0,1], ['A', 'B', 'C'])
plt.yticks([-0.4,-0.2, 0, 0.2, 0.4])
plt.minorticks_on()
plt.grid(1)
plt.show()
```

Изобразим сразу два графика функций:

1. $y=x(x+1)(x-1)$
2. $z=e^{0.5x}-1$

```python
X = np.linspace(-1., 1., 17)
Y = X * (X+1) * (X-1)
Z = np.exp(0.5*X) - 1
plt.plot(X, Y, marker='o', label=r'$y=x(x+1)(x-1$')
plt.plot(X, Z, marker='s', label=r'$y=e^{x/2}-1$')
plt.title('Графики функций')
plt.xlabel('X')
plt.ylabel('Y')
plt.minorticks_on()
plt.grid(1)
plt.legend()
plt.show()
```

```python
X = np.linspace(-1., 1., 101)
Y = X * (X+1) * (X-1)
Z = np.exp(0.5*X) - 1
plt.plot(X, Y, label=r'$y=x(x+1)(x-1$')
plt.plot(X, Z, label=r'$y=e^{x/2}-1$')
plt.title('Графики функций')
plt.xlabel('X')
plt.ylabel('Y')
plt.minorticks_on()
plt.grid(1)
plt.legend()
plt.show()
```

```python
X = np.linspace(-1., 1., 100)
Y = X * (X+1) * (X-1)
Z = np.exp(0.5*X) - 1
plt.plot(X, Y, linewidth='4.0', label=r'$y=x(x+1)(x-1$')
plt.plot(X, Z, linestyle='--', label=r'$y=e^{x/2}-1$')
plt.title('Графики функций')
plt.xlabel('X')
plt.ylabel('Y')
plt.minorticks_on()
plt.grid(1)
plt.legend()
plt.show()
```

## Bars

```python
R = [1., 10.,-2.,5.,2.]
plt.bar(range(5), R, color='LightGreen', edgecolor='DarkGreen')
plt.minorticks_on()
plt.show()
```

```python
plt.bar(range(5), R, color='LightGreen')
for i in range(5):
    plt.text(i, R[i], str(R[i]))
plt.minorticks_on()
plt.show()
```

```python
plt.barh(range(5), R, color='LightGreen')
for i in range(5):
    plt.text(R[i], i, str(R[i]))
plt.minorticks_on()
plt.show()
```

```python
plt.figure(figsize=(10,5))
R1 = 5*np.random.random(10)
R2 = R1 + 2*np.random.random(10)
R3 = R2 + 4*np.random.random(10)
bar_pos = np.arange(10)
plt.bar(bar_pos, R1, align='edge', width=0.24, label='R1')
plt.bar(bar_pos+0.25, R2, align='edge', width=0.24, label='R2')
plt.bar(bar_pos+0.50, R3, align='edge', width=0.24, label='R3')
plt.legend(loc='best')
plt.xticks(range(10))
plt.ylim(0,10)
for i in range(10):
    t1 = plt.text(bar_pos[i], R1[i], "%.2f" % R1[i], verticalalignment='bottom')
    t1.set_rotation('vertical')
    t2 = plt.text(bar_pos[i]+0.25, R2[i], "%.2f" % R2[i], verticalalignment='bottom')
    t2.set_rotation('vertical')
    t3 = plt.text(bar_pos[i]+0.50, R3[i], "%.2f" % R3[i], verticalalignment='bottom')
    t3.set_rotation('vertical')
plt.show()
```

Изобразим уровни следующей функции на плоскости:
$$
z=e^{-\sqrt{x^2+y^2}}\cos\left(4\pi\sqrt{x^2+y^2}\right)
$$

```python
RR = np.sqrt(XX**2+YY**2)
ZZ = np.exp(-RR)*np.cos(4*np.pi*RR)
```

```python
im = plt.imshow(ZZ, cmap=plt.cm.Greys)
plt.colorbar(im)
plt.show()
```

```python
im = plt.imshow(ZZ, cmap=plt.cm.Greys, origin='lower', extent=(-1.,1.,-2.,2.))
plt.colorbar(im)
plt.show()
```

```python
im = plt.imshow(ZZ, cmap=plt.cm.Greys, origin='lower', extent=(-1.,1.,-2.,2.), aspect='auto')
plt.colorbar(im)
plt.show()
```

## Contour

```python
X=np.linspace(-2,2,101)
Y=np.linspace(-2,2,101)
XX, YY = np.meshgrid(X, Y)
```

$$
z = e^{-(x-1)^2-(y-1)^2} - e^{-(x+1)^2-(y+1)^2}
$$

```python
ZZ = np.exp(-(XX-1)**2-(YY-1)**2) - np.exp(-(XX+1)**2-(YY+1)**2)
```

```python
plt.figure(figsize=(7,6))
zlevels = np.linspace(ZZ.min(), ZZ.max(), 12)
plt.contour(ZZ, levels=zlevels, extent=(-2.,2.,-2.,2.))
plt.show()
```

```python
zlevels = np.linspace(ZZ.min(), ZZ.max(), 12)
plt.figure(figsize=(7,6))
cp = plt.contour(ZZ, levels=zlevels, extent=(-2.,2.,-2.,2.))
plt.clabel(cp, cp.levels, colors='k')
# plt.colorbar(cp)
plt.show()
```

```python
zlevels = np.linspace(ZZ.min(), ZZ.max(), 10)
plt.figure(figsize=(7,6))
cp = plt.contourf(ZZ, levels=zlevels, extent=(-2.,2.,-2.,2.))
plt.colorbar(cp)
plt.show()
```

```python
zlevels = np.linspace(ZZ.min(), ZZ.max(), 10)
plt.figure(figsize=(7,6))
cp=plt.contour(ZZ, levels=zlevels, extent=(-1.,1.,-2.,2.), colors='k', linewidths=0.75)
plt.contourf(ZZ, levels=zlevels, extent=(-1.,1.,-2.,2.))
plt.clabel(cp, colors='k', fmt='%.2f')
plt.show()
```

## Scatter 3D

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.random.random(30)
Y = np.random.random(30)
Z = np.random.random(30)
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.gca(projection='3d')
R = np.sqrt(X**2+Y**2+Z**2)
ax.scatter3D(X, Y, Z, s=400*R, c=R, edgecolors='k', alpha=0.75)
plt.show()
```

## Plot3D

```python
T = np.linspace(0, 30, 1500)
X = np.sin(2*np.pi*T)
Y = np.sin(3*np.pi*T)
Z = np.sin(5*np.pi*T)

fig = plt.figure(figsize=(10,8))
ax = fig.gca(projection='3d')
ax.plot3D(X, Y, Z, linewidth=2.0)
plt.show()
```

## Surface 3D

```python
X=np.linspace(-4,4,100)
Y=np.linspace(-4,4,100)
XX, YY = np.meshgrid(X, Y)
ZZ = np.exp(-(XX-1)**2-(YY-1)**2) - np.exp(-(XX+1)**2-(YY+1)**2)

fig = plt.figure(figsize=(7,5))
ax = fig.gca(projection='3d')
ax.plot_surface(XX, YY, ZZ, color='LightGreen')
ax.azim=-45
plt.show()
```

## Contour 3D

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.contour3D(XX, YY, ZZ, levels=np.linspace(ZZ.min(), ZZ.max(), 30))
ax.azim=-60
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
ax.contourf3D(XX, YY, ZZ, levels=np.linspace(ZZ.min(), ZZ.max(), 12))
ax.azim=-60
plt.show()
```
