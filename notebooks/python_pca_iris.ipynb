```python
import numpy as np
import matplotlib.pyplot as plt
# import mltools.funcs as funcs
import mltools.pca as pca
import mltools.location_scatter as location
# import mltools.plots as plots
import sklearn.datasets as datasets
```

```python
X, Y = datasets.load_iris(return_X_y=True)
```

```python
X0 = X
Y0 = Y
```

```python
c = location.location(X0)
print(c)
```

```python
X1 = X0 - c
```

```python
a1, L1 = pca.find_pc(X1)
U1 = pca.project_line(X1, a1)
print(L1, a1)
```

```python
X2 = pca.project(X1, a1)
```

```python
a2, L2 = pca.find_pc(X2)
U2 = pca.project_line(X2, a2)
print(L2, a2)
```

```python
plt.scatter(U1, U2, c=Y0, edgecolors='k', s=25)
plt.show()
```

```python
X3 = pca.project(X2, a2)
```

```python
a3, L3 = pca.find_pc(X3)
U3 = pca.project_line(X3, a3)
print(L3, a3)
```

```python
X4 = pca.project(X3, a3)
```

```python
a4, L4 = pca.find_pc(X4)
U4 = pca.project_line(X4, a4)
print(L4, a4)
```

```python
plt.bar(range(4), [L1, L2, L3, L4])
plt.minorticks_on()
plt.show()
```

```python
V = np.array([L1,L2,L3,L4])
```

```python
plt.bar(range(4), V/V.sum())
plt.minorticks_on()
plt.show()
```
