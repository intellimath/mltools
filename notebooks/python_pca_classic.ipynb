```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.decomposition as decomp
```

```python
df = pd.read_csv("data/starsCYG.csv", sep=',')
```

```python
X1 = df["log.Te"].values
X2 = df["log.light"].values
X = np.c_[X1, X2]
print(X)
```

```python
plt.scatter(X1, X2, s=64, c='w', edgecolors='k')
plt.show()
```

```python
c1 = np.mean(X1)
c2 = np.mean(X2)
c = np.array([c1,c2])
```

```python
plt.scatter(X1, X2, s=64, c='w', edgecolors='k')
plt.scatter([c1], [c2], c='r')
plt.show()
```

```python
S = np.dot(X.T, X) / len(X)
print(S)
```

```python
np.linalg.eigvals(S)
```

```python
V, A = np.linalg.eig(S)
```

```python
plt.scatter(X1, X2, s=64, c='w', edgecolors='k')
plt.scatter([c1], [c2], c='r')
plt.plot([c1, c1+A[0,0]], [c2, c2+A[1,0]])
plt.plot([c1, c1+A[0,1]], [c2, c2+A[1,1]])
plt.show()
```
