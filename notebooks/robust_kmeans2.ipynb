```python
import sys
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn import cluster

import mltools.kmeans as kmeans
import mltools.aggfuncs as aggfuncs
import mltools.funcs as funcs
```

```python
S=100.
X_1, Y_1 = make_blobs(n_samples=200, n_features=2, 
                    centers=[(-30,20), (30,-30)], 
                    cluster_std=[5.0, 5.0])
Y_1 += 1

X_2, Y_2 = make_blobs(n_samples=100, n_features=2, 
                    centers=[(10,40),(40, 70)], 
                    cluster_std=[30., 30.])
Y_2 = 3
print(X_1.shape, X_2.shape)
X1 = np.vstack([X_1, X_2])
Y1 = np.hstack([Y_1, Y_2])
X1 /= S

plt.scatter(X1[:,0], X1[:,1], s=16, cmap=plt.cm.rainbow)
plt.show()
```

```python
X_1, Y_1 = make_blobs(n_samples=200, n_features=2, 
                    centers=[(-30,30), (30,-30)], 
                    cluster_std=[5.0, 5.0])
Y_1 += 1

X_2, Y_2 = make_blobs(n_samples=200, n_features=2, 
                    centers=[(0,70),(60, 70)], 
                    cluster_std=[40., 40.])
Y_2 = 3
print(X_1.shape, X_2.shape)
X2 = np.vstack([X_1, X_2])
Y2 = np.hstack([Y_1, Y_2])
X2 /= S

plt.scatter(X2[:,0], X2[:,1], s=16, cmap=plt.cm.rainbow)
plt.show()
```

```python
xmin, xmax = -150./S, 170./S
ymin, ymax = -50./S, 170./S
def draw_map(cls, i):
    x_range = np.linspace(xmin, xmax, 100)
    y_range = np.linspace(ymin, ymax, 100)
    XX, YY = np.meshgrid(x_range, y_range)
    XY = np.c_[XX.ravel(), YY.ravel()]
    ZZ = cls(XY)
    ZZ = ZZ[:,i]
    ZZ = ZZ.reshape(XX.shape)
    im = plt.imshow(ZZ, origin='bottom', extent=(xmin,xmax,ymin,ymax), aspect="auto")
#     plt.colorbar(im)
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
```

```python
agg = aggfuncs.WMMean(funcs.SoftQuantileFunc(funcs.SoftAbs(0.001), 0.5))
rkm = kmeans.RKMeansMahalanobis(2, agg, n_iter_c=100, n_iter=500)
rkm.fit(X2)
```

```python
print(rkm.c)
print(rkm.S1)
```

```python
plt.plot(rkm.qvals)
```

```python
plt.scatter(X2[:,0], X2[:,1], s=25, c='w', edgecolors='k', linewidths=0.3)
plt.scatter(rkm.c[:,0], rkm.c[:,1], s=64, c='k')

XX, YY = np.meshgrid(np.linspace(-1.5, 2.0, 100), np.linspace(-1.0, 2.0, 100))
XY = np.c_[XX.ravel(), YY.ravel()]
Z = np.array([rkm.norm2(xy) for xy in XY])
Z = Z.reshape(XX.shape)
plt.contour(Z, extent=(-1.5, 2.0, -1.0, 2.0), levels=np.linspace(Z.min(), Z.max(), 30))

plt.show()
```
